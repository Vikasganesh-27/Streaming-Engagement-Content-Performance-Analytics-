**RENAME COLUMNS**

```python
df = df.rename(columns={
    "artist(s)_name": "artist_name",
    "released_year": "release_year",
    "released_month": "release_month",
    "released_day": "release_day",
    "in_spotify_playlists": "spotify_playlists",
    "in_spotify_charts": "spotify_charts",
    "in_apple_playlists": "apple_playlists",
    "in_apple_charts": "apple_charts",
    "in_deezer_playlists": "deezer_playlists",
    "in_deezer_charts": "deezer_charts",
    "in_shazam_charts": "shazam_charts",
    "danceability_%": "danceability",
    "valence_%": "valence",
    "energy_%": "energy",
    "acousticness_%": "acousticness",
    "instrumentalness_%": "instrumentalness",
    "liveness_%": "liveness",
    "speechiness_%": "speechiness"
})
```

**RELEASE DATE**

```python
df['release_date'] = pd.to_datetime(
    df[['release_year', 'release_month', 'release_day']].astype(str).agg('-'.join, axis=1),
    errors='coerce'
)
```

```python
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')
df['deezer_playlists'] = pd.to_numeric(df['deezer_playlists'], errors='coerce')
df['shazam_charts'] = pd.to_numeric(df['shazam_charts'], errors='coerce')
```

**Dimension and Fact Tables**

```python
dim_artist = df[['artist_name', 'artist_count']].drop_duplicates().reset_index(drop=True)
dim_artist['artist_id'] = range(1, len(dim_artist)+1)
dim_artist.to_csv("dim_artist.csv", index=False)
```

```python
# Merge artist_id from dim_artist
df = df.merge(dim_artist[['artist_id', 'artist_name']], on='artist_name', how='left')

# Create dim_track
dim_track = df[['track_name', 'artist_id', 'release_date', 'bpm', 'key', 'mode',
                'danceability', 'valence', 'energy', 'acousticness', 'instrumentalness',
                'liveness', 'speechiness', 'cover_url']].drop_duplicates().reset_index(drop=True)

# Assign unique track_id
dim_track['track_id'] = range(1, len(dim_track)+1)

# Preview
print(dim_track.head())
dim_track.to_csv("dim_track.csv", index=False)
```

```python
dim_track.head()
```

```python
# Merge track_id from dim_track
df = df.merge(dim_track[['track_id', 'track_name', 'artist_id']], 
              on=['track_name', 'artist_id'], how='left')
```

```python
# Create fact_streams
fact_streams = df[['track_id', 'spotify_playlists', 'spotify_charts',
                   'apple_playlists', 'apple_charts', 'deezer_playlists', 
                   'deezer_charts', 'shazam_charts', 'streams']].drop_duplicates().reset_index(drop=True)

# Preview
fact_streams.head()
fact_streams.to_csv("fact_streams.csv", index=False)
```

# ANALYSIS & METRICS

**Top Tracks by Total Streams**

```python

df = pd.merge(fact_streams, dim_track, on='track_id', how='inner')

track_streams = df.groupby(['track_id', 'track_name'], as_index=False)['streams'].sum()


track_streams = track_streams.sort_values(by='streams', ascending=False)
track_streams.head(10)
```

**Top Artists by Streams**

```python

df = pd.merge(fact_streams, dim_track, on='track_id', how='inner')
df = pd.merge(df, dim_artist, on='artist_id', how='inner')
artist_streams = df.groupby(['artist_id', 'artist_name'], as_index=False)['streams'].sum()
artist_streams = artist_streams.sort_values(by='streams', ascending=False)
artist_streams.head(10)
```

**Streams by Release Year**

```python
# Ensure release_year exists
track_data = dim_track.merge(fact_streams, on='track_id', how='left')
track_data['release_year'] = track_data['release_date'].dt.year

# Total streams per release year
streams_by_year = track_data.groupby('release_year')['streams'].sum().reset_index().sort_values(by='release_year')

streams_by_year.head(10)
```

**Engagement Score**

```python
import pandas as pd

# Merge fact_streams with dim_track
df = pd.merge(fact_streams, dim_track, on='track_id', how='inner')

df['engagement_score'] = (
    df['streams'] * 0.6 +
    (df['spotify_playlists'] + df['apple_playlists'] + df['deezer_playlists']) * 0.25 +
    (df['spotify_charts'] + df['apple_charts'] + df['deezer_charts'] + df['shazam_charts']) * 0.15
)

engagement_df = df[['track_id', 'track_name', 'engagement_score']]

engagement_df = engagement_df.sort_values(by='engagement_score', ascending=False)
pd.options.display.float_format = '{:,.2f}'.format
engagement_df.head(10)
```

**Total Playlist Exposure per Track**

```python

df = pd.merge(fact_streams, dim_track, on='track_id', how='inner')

df['total_playlist_exposure'] = (
    df['spotify_playlists'] + df['apple_playlists'] + df['deezer_playlists']
)

playlist_df = df[['track_id', 'track_name', 'total_playlist_exposure']]

playlist_df = playlist_df.sort_values(by='total_playlist_exposure', ascending=False)
playlist_df.head(10)

```

**Total Chart Exposure per Track**

```python

df = pd.merge(fact_streams, dim_track, on='track_id', how='inner')
df['total_chart_exposure'] = (
    df['spotify_charts'] + df['apple_charts'] + df['deezer_charts'] + df['shazam_charts']
)
chart_df = df[['track_id', 'track_name', 'total_chart_exposure']]
chart_df = chart_df.sort_values(by='total_chart_exposure', ascending=False)
chart_df.head(10)
```

**Top Tracks per Platform**

```python

df = pd.merge(fact_streams, dim_track, on='track_id', how='inner')

spotify_df = df.groupby(['track_id', 'track_name'], as_index=False)['spotify_playlists'].sum()

spotify_df = spotify_df.rename(columns={'spotify_playlists': 'spotify_playlist'})

spotify_df = spotify_df.sort_values(by='spotify_playlist', ascending=False)
spotify_df.head(10)
```

**Stream Efficiency Score**

```python

df = pd.merge(fact_streams, dim_track, on='track_id', how='inner')

df['total_exposure'] = (
    df['spotify_playlists'] + df['apple_playlists'] + df['deezer_playlists'] +
    df['spotify_charts'] + df['apple_charts'] + df['deezer_charts'] + df['shazam_charts']
)

df['total_exposure'] = df['total_exposure'].replace(0, np.nan)

df['efficiency_score'] = df['streams'] / df['total_exposure']

efficiency_df = df[['track_id', 'track_name', 'efficiency_score']]

efficiency_df = efficiency_df.sort_values(by='efficiency_score', ascending=False)
efficiency_df.head(10)
```

**Platform ROI**

```python

df = pd.merge(fact_streams, dim_track, on='track_id', how='inner')

df['spotify_playlists'] = df['spotify_playlists'].replace(0, np.nan)
df['apple_playlists'] = df['apple_playlists'].replace(0, np.nan)
df['deezer_playlists'] = df['deezer_playlists'].replace(0, np.nan)

df['spotify_roi'] = df['streams'] / df['spotify_playlists']
df['apple_roi'] = df['streams'] / df['apple_playlists']
df['deezer_roi'] = df['streams'] / df['deezer_playlists']

roi_df = df[['track_id', 'track_name', 'spotify_roi', 'apple_roi', 'deezer_roi']]

roi_df = roi_df.sort_values(by='spotify_roi', ascending=False)
roi_df.head(10)
```

**Audio Feature Average per Artist**

```python

df = pd.merge(dim_track, dim_artist, on='artist_id', how='inner')

audio_features = df.groupby('artist_name', as_index=False).agg({
    'danceability': 'mean',
    'energy': 'mean',
    'valence': 'mean',
    'bpm': 'mean'
})

audio_features = audio_features.sort_values(by='artist_name')
audio_features.head(10)
```

# USER-LEVEL / EVENT-LEVEL ANALYTICS

**User data**

```python
np.random.seed(42)

num_users = 10000  # reduced size

countries = ["US", "UK", "CA", "IN", "DE", "BR", "AU"]
country_weights = [0.25, 0.10, 0.08, 0.30, 0.07, 0.12, 0.08]

genders = ["Male", "Female", "Other"]
gender_weights = [0.48, 0.48, 0.04]

subscription_types = ["Free", "Premium"]
subscription_weights = [0.65, 0.35]

dim_user = pd.DataFrame({
    "user_id": range(1, num_users + 1),
    "country": np.random.choice(countries, size=num_users, p=country_weights),
    "age": np.random.randint(18, 60, size=num_users),
    "gender": np.random.choice(genders, size=num_users, p=gender_weights),
    "subscription_type": np.random.choice(subscription_types, size=num_users, p=subscription_weights),
    "signup_date": pd.to_datetime("2020-01-01") + pd.to_timedelta(
        np.random.randint(0, 5*365, size=num_users), unit="D")
})

dim_user.head()
dim_user.to_csv("dim_user.csv", index=False)
```

```python
dim_user.info()
```

**Fact Listening Events**

```python
np.random.seed(42)

# Reduce users to 1000
dim_user_small = dim_user.sample(1000, random_state=42).sort_values("user_id")

# 5–15 events per user
events_per_user = np.random.randint(5, 15, size=1000)

event_rows = []
track_ids = dim_track["track_id"].tolist()
devices = ["Mobile", "Desktop", "Smart Speaker"]
device_weights = [0.65, 0.25, 0.10]

current_event_id = 1

for user_id, count in zip(dim_user_small["user_id"], events_per_user):
    for _ in range(count):
        track = np.random.choice(track_ids)
        session_length = np.random.randint(30, 600)
        listened = np.random.randint(10, session_length)

        event_rows.append([
            current_event_id,
            user_id,
            track,
            pd.to_datetime("2020-01-01") + pd.to_timedelta(
                np.random.randint(0, 5*365), unit="D"),
            np.random.choice(devices, p=device_weights),
            session_length,
            listened,
            1 if listened >= session_length * 0.8 else 0
        ])
        current_event_id += 1

fact_listening_events = pd.DataFrame(event_rows, columns=[
    "event_id", "user_id", "track_id", "timestamp",
    "device", "session_length_sec", "listened_seconds", "completed"
])

fact_listening_events.to_csv("fact_listening_events.csv", index=False)
```

**Events by Device**

```python
events_by_device = (
    fact_listening_events
        .groupby('device')
        .size()
        .reset_index(name='event_count')
        .sort_values('event_count', ascending=False)
)

print(events_by_device)
```

**Average Listening Time per User**

```python
total_listened_seconds = fact_listening_events['listened_seconds'].sum()
total_users = len(dim_user)

avg_listening_minutes_per_user = round(total_listened_seconds / total_users / 60, 2)
print(avg_listening_minutes_per_user)
```

**Completion Rate per User**

```python
# Count total sessions and compute completion rate for all users
completion_rate_per_user = (
    dim_user[['user_id']]
    .merge(
        fact_listening_events.groupby('user_id')['completed']
            .agg(total_sessions='count', completed_sum='sum')
            .reset_index(),
        on='user_id',
        how='left'
    )
)

# Fill NaN for users with zero events
completion_rate_per_user['total_sessions'] = completion_rate_per_user['total_sessions'].fillna(0)
completion_rate_per_user['completed_sum'] = completion_rate_per_user['completed_sum'].fillna(0)

# Calculate completion rate percentage
completion_rate_per_user['completion_rate_percent'] = (
    completion_rate_per_user['completed_sum'] / completion_rate_per_user['total_sessions'].replace(0,1) * 100
)

completion_rate_per_user = completion_rate_per_user[['user_id', 'total_sessions', 'completion_rate_percent']].sort_values(by = 'completion_rate_percent', ascending = False)
print(completion_rate_per_user.head(10))
```
